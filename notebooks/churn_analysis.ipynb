```python
df = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn[1].csv')
```

```python
# Section 3: missing values and blanks
# Note: TotalCharges may contain blanks that read as empty strings
df.replace(' ', np.nan, inplace=True)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.isnull().sum()
```

```python
# Section 4: churn distribution
print(df['Churn'].value_counts())
sns.countplot(data=df, x='Churn')
```

```python
# Section 5: driver analyses
# Tenure vs Churn
display(df.groupby('Churn')['tenure'].describe())
sns.boxplot(data=df, x='Churn', y='tenure')

# Contract vs Churn
display(pd.crosstab(df['Contract'], df['Churn'], margins=True))
(pd.crosstab(df['Contract'], df['Churn'], normalize='index')*100).round(2)

# Payment Method vs Churn
display(pd.crosstab(df['PaymentMethod'], df['Churn'], margins=True))

# Internet Service vs Churn
display(pd.crosstab(df['InternetService'], df['Churn'], margins=True))
```
